```python
import scorchgeo
```

```python
m = scorchgeo.Map()
url = "https://github.com/opengeos/datasets/releases/download/us/us_states.geojson"
m.add_geojson(url)
m
```

```python
m = scorchgeo.Map()
url = "https://github.com/opengeos/datasets/releases/download/us/us_states.zip"
m.add_shp(url, name="US States")
m
```

```python
m = scorchgeo.Map()
url = "https://github.com/opengeos/datasets/releases/download/us/us_states.zip"
m.add_vector(url, name="US States")
m.add_basemap()
m
```

```python
m = scorchgeo.Map()
tile_layer = "https://github.com/opengeos/datasets/releases/download/raster/dem_90m.tif"
m.add_raster(tile_layer)
m
```

```python
m = scorchgeo.Map()
url = "https://i.imgur.com/8YGYanS.jpeg"
m.add_image(url, bounds=[[-90, -180], [90, 180]])
m
```

```python
m = scorchgeo.Map()
url = "https://data.opengeos.org/patricia_nasa.webm"
bounds = [
    [-130, 32],
    [-100, 32],
    [-100, 13],
    [-130, 13],
]
m.add_video(url, bounds)
m
```
