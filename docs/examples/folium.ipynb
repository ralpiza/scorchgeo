```python
import scorchgeo.foliumap as scorchgeo
```

```python
m = scorchgeo.Map(center=[20, 0], zoom=2)
url = "https://github.com/opengeos/datasets/releases/download/us/us_states.geojson"
m.add_geojson(url, name="US States")
m.add_layer_control()
m
```
